```python
!pip -q install osmnx networkx

# Pickle 生成
import osmnx as ox, pickle, pathlib


out_path = pathlib.Path('/content/drive/MyDrive/Toyohashi_Qaneering/toyohashi_drive_graph.pkl')

place = {"city": "Toyohashi", "state": "Aichi", "country": "Japan"}
G = ox.graph_from_place(place, network_type="drive")    # ここはオンラインで OSM を取得

out_path.parent.mkdir(parents=True, exist_ok=True)       # ディレクトリが無ければ作成
with open(out_path, "wb") as f:
    pickle.dump(G, f, protocol=pickle.HIGHEST_PROTOCOL)

print(f"Saved to: {out_path}")
```
